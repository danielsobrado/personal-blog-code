# Entropy

```python
from scipy import stats
```

```python
print(stats.entropy(pk=[0.95, 0.05]))
```

```python
print(stats.entropy(pk=[0.95, 0.05],base=2))
```

```python
import numpy as np
pk = [0.95, 0.05]
print(-np.sum(pk * np.log(pk)/np.log(2), axis=0))
```

```python
import math
print(-0.95 * math.log(0.95,2) + -0.05 * math.log(0.05,2))
```

# Kullback Leibler Divergence

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x_axis = np.arange(-10, 10, 0.25)
dist_a = stats.norm.pdf(x_axis,0,25)
dist_b = stats.norm.pdf(x_axis,1,25)
plt.plot(x_axis, dist_a)
plt.plot(x_axis, dist_b)
plt.fill_between(x_axis, dist_a, dist_b, where=dist_b>=dist_a, facecolor='cyan', interpolate=True)
plt.fill_between(x_axis, dist_a, dist_b, where=dist_b<=dist_a, facecolor='magenta', interpolate=True)
plt.show()
```
